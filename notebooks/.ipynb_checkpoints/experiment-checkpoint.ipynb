```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
real_path = "../data/raw/real_profiles"
fake_path = "../data/raw/fake_profiles"

print("Real images:", len(os.listdir(real_path)))
print("Fake images:", len(os.listdir(fake_path)))
```

```python
real_images = os.listdir(real_path)
fake_images = os.listdir(fake_path)

print("Sample real image:", real_images[0])
print("Sample fake image:", fake_images[0])
```

```python
img = Image.open(os.path.join(real_path, real_images[0]))
plt.imshow(img)
plt.axis("off")
plt.title("Real Profile")
```

```python
img = Image.open(os.path.join(fake_path, fake_images[0]))
plt.imshow(img)
plt.axis("off")
plt.title("Fake Profile")
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
IMG_SIZE = (224, 224)
BATCH_SIZE = 32
```

```python
dataset = tf.keras.preprocessing.image_dataset_from_directory(
    "../data/raw",
    labels="inferred",
    label_mode="binary",
    batch_size=BATCH_SIZE,
    image_size=IMG_SIZE,
    shuffle=True
)
```

```python
train_size = int(0.8 * len(dataset))

train_dataset = dataset.take(train_size)
val_dataset = dataset.skip(train_size)
```

```python
for images, labels in train_dataset.take(1):
    print(images.shape)
    print(labels.shape)
```

```python
normalization_layer = layers.Rescaling(1./255)

train_dataset = train_dataset.map(lambda x, y: (normalization_layer(x), y))
val_dataset = val_dataset.map(lambda x, y: (normalization_layer(x), y))
```

```python
base_model = tf.keras.applications.MobileNetV2(
    input_shape=(224, 224, 3),
    include_top=False,
    weights="imagenet"
)
```

```python
base_model.trainable = False
```

```python
model = keras.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation="relu"),
    layers.Dropout(0.3),
    layers.Dense(1, activation="sigmoid")
])
```

```python
model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
```

```python
model.summary()
```

```python
history = model.fit(
    train_dataset,
    validation_data=val_dataset,
    epochs=5
)
```

```python
plt.plot(history.history['accuracy'], label='train_accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')

plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.title("Training vs Validation Accuracy")
plt.show()
```

```python
plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')

plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.title("Training vs Validation Loss")
plt.show()
```

```python
def predict_image(path):

    img = Image.open(path).resize((224,224))
    img_array = np.array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    prediction = model.predict(img_array)[0][0]

    print("Prediction score:", prediction)

    if prediction > 0.5:
        print("Real Profile")
    else:
        print("Fake Profile")

    plt.imshow(img)
    plt.axis("off")
```

```python
predict_image("../data/raw/real_profiles/000001.jpg")
```

```python
predict_image("../data/raw/fake_profiles/" + fake_images[0])
```
